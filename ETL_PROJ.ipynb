```python
 url = 'http://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?class=1;id=2018;type=year'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all("tr",class_="data1")
results
```

```python
for result in results:
    name = result.find("td").text
    
    print(name)
  
    
```

```python
for result in results:
    name = result.find("td").text
    matches = result.find("td",  class_="padAst").text
   
    print(name)
    print(matches)

    
```
